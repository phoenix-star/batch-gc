Batch Script Garbage Collection
===============================

I decided to implement simple garbage collection for envrionment variables, with Windows batch scripting. The results are less than stellar. I quickly realized that it's easy to write batch scripts; it's not easy to write scripts well. Sure this could have been done in Powershell, BASH, or, any other shell language. Would have been more sane. But, where's the fun in that? :)

For source code, check out https://github.com/phoenix-star/batch-gc

Big thanks to the following:

- Plethora of StackOverflow posts on batch scripting
- [SS64 CMD Entries](http://ss64.com/nt/)
- ConEmu-Maximus5 developers and maintainers, allowing me to have a sane development environment

---

<a name="what"></a>
What does it do?
----------------

batch-gc is a collection of batch scripts that allow for references, reference counting and garbage collection in a CMD shell/environment. Rather than create environment variables with values, the scripts allow for creating envrionment variables with references that can be shared and deleted, marked as garbage. References, and other values necessary for the logic are stored in files rather than the environment space. This was due to some design decisions, but the unintended consequence is that users get very few perks of using references without adding more variables into the environment space.

<a name="why"></a>
Why?
----

I wanted to challenge myself to write well-written batch scripts.

Cleanliness
-----------

   The overall format of the code should just look nice. Python lends itself to writing code that just-looks-good. That's what I

<a name="design_goals-readibility"></a>
Readibility
-----------

   There are some powerful things that can be done in batch scriptin. They can be terse, and when they are, they read like   
   heiroglyphics. And when they aren't terse, well, yeah. This makes it difficult to follow code, which isn't desireable if you want to share 
   your work with others. So, I aimed for verbosity for the sake of clarity. Performance be damned.

<div style="text-align:right;">[Back to top](#bc)</div>

---

<a name="design_goals-consistency"></a>
Consistency
-----------

   There are many ways to be inconsistent with batch scripting. To name a few:
   
   * Executing sections of code within a script:
    * Simply falling into the next section
    * `GOTO` statements 
    * `CALL` statements to other sections
       
   * Executing code within an external script:
    * `CALL` statements
    * `CMD` statements
    * `START` statements 
       
   * Returning values:
    * `ECHO` and redirection
    * Global variables 
    * Throwing variables over the wall with `ENDLOCAL`
    * Writing to files
   
Each script has an identical layout. Each script is treated like a single function. I avoid `GOTO` statements. Variables passed subscribe  
to a naming convention. Variables passed out subscribe to a naming convention. Initializations involving parameters always occur within the 
same section. Even just the fact that I decided to go with everything is capitalized, since batch scripts, other than string comparisons, 
are case-insensitive.
   
I'm sure this list is exhaustive.

<a name="layout-init-setlocal"></a>
###_`SETLOCAL ENABLEDELAYEDEXPANSION`_

This declaration does two things

1\. It forces all following variable declarations with the `SET` command to be local. If a variable existed prior to this delcaration, it may  
be read, but it cannot be overwritten. This allows for the behavior observed in functions in most programming languages: variables declared within a function are not accessible outside of the function. In fact, `SETLOCAL` can be nested allowing different levels of scope, just how nested function calls, or declarations, create new levels of scope.

    
2\. Usually, environment variables are expanded when they are parsed, not executed. This causes %myvariable%, anywhere in a script, to be immediately replaced by its current value. This has nasty consequences when dealing with varibles in loops, or variables that are set multiple times in a script. The option `EnableDelayedExpansion` remedies this by expanding a variable when executed, so long as it's wrapped in `!`.

Every script has this line so that I don't pollute or corrupt the current namespace, and so that delayed variable expansion is available by default.

<div style="text-align:right;">[Back to top](#bc)</div>

---

<a name="layout-init-base_dir"></a>
###_`SET BASE_DIR=%~DP0`_

`%0` Is always the absolute path the script being invoked. The inclusion of `~` allows for the inclusion of options that alter the return value of `%0`. `~D` returns the only the drive letter in the path, while `~P` returns only the path up to the file, minus the drive letter, file name and file extension. These options combined into `~DP` will return the base directory of the script being invoked.

Why would I want that? Files required for reading and writing global variables are saved in the projects base directory. `BASE_DIR` seemed like the safest place to dump these files.

<div style="text-align:right;">[Back to top](#bc)</div>

---

<a name="layout-init-arguments"></a>
###*`SET IN_, OUT_, INOUT_`*

The remainder of the section is reserved exclusively for initializing variables using arguments passed to the script. If the value passed in is to be used for internal logic, it's name is prepended with `IN_`; if the value passed in is meant for storing a return value, its prepended with `OUT_`; those intended for both are prepended with `INOUT_`. After the `_`, the names are pretty straight forward. If it's the name of an external variable, it's usually `VAR_NAME`. If it's something to be treated as a value, it's usually `VALUE`. The variable names should spell out what's expected.

These variables are also set in the order of the positional arguments. So the first argument, `%1`, is always set before the second argument `%2`, which is always set, and, yada yada.

Most importantly, this section is reserved for initializing variables using arguments passed to a script. It is **never** used for initializing other variables. So if you ever want to see what arguments these scripts expect, this is the only section you should have to read.

<a name="layout-return"></a>
`:RETURN`
---------

I wanted it blantatly obvious when a value was being expliclitly returned. This section is reserved for `OUT_` or `INOUT_` variables. The process of "returning" variables allows for passing values over the scope boundary. 

For this project, when a script is expected to return a value, a name for the variable to hold the return value is passed as an argument and internally held as a `OUT_` or `INOUT_` variable. At the end of the script, a variable, with the provided name, is injected into the outer scope, with the return value. So, you can choose the name of your return variable, rathr than the other way around.

<a name="layout-return-endlocal"></a>
###_`ENDLOCAL`_

This command ends the local scope, making it possible to return values. It also ends delayed variable expansion. Thus, from this point forward, only `OUT_` and `INOUT_` are altered using `SET`.

<a name="layout-return-variables"></a>
###_`SET %OUT_VAR_1%=%VAR_N%`_

If there are values to be returned, the entire section looks something like this:

```dos
:RETURN
    ENDLOCAL & (
        SET %OUT_VAR_1%=%VAR_1%
        SET %OUT_VAR_2%=%VAR_2%
        ...
        SET %OUT_VAR_N%=%VAR_N%
    )
```

If there are no values to be returned, the entire section is identical to `:RETURN` in the template. Every `OUT_` variable is initialized in `:INITILIZATION`. It has a corresponding environment variable that is used internally for holding the return value.

Similar to the `:INITIALIZATION`, the section is reserved for nothing else other than returning values.

<a name="refs_gc-references"></a>
References
----------
In short, a reference is a way of uniquely identifying some _thing_. What this _thing_ is, isn't as important. I'll just refer to it as $Thing$ from now on. 

If you ever printed a reference to the console, it usually looks a combination of hex characters and some other info. C++, Java, Python and other languages rely on references for multiple things.

<div style="text-align:right;">[Back to top](#bc)</div>

---

<a name="refs_gc-references-equality"></a>
###_Equality_
I have two variables, `a` and `b`. How do I know if they are equal? Via `Reference`s.

To trivialize, a `Reference` is as an ID. No two `Reference`s may be equal. Variables can be equal to references, but to get a point across, I'm going to say that variables point to a `Reference`.

$$Variable\ =>\ Reference$$

Different variables can point to the same `Reference` and that is perfectly valid. What's invalid is multiple `Reference`s being identical. 

Let's say I have a $Reference$, $X_{Refrence}$, and two variables, $A_{Variable}$ and $B_{Variable}$. $A_{Variable}$ points to $X_{Reference}$ and $B_{Variable}$ point to $X_{Reference}$.

$$A_{Variable}\ =>\ X_{Reference}$$
$$B_{Variable}\ =>\ X_{Reference}$$

Apply the transitive property, and you're done. As far as equality is concerend, if `a` and `b` hold the same `Reference`, `a` and `b` are the same `Reference` and `a` and `b` are equal. From this, you can deduce that if `a` and `b` do not hold the same `Reference`, `a` and `b` are not equal.

<div style="text-align:right;">[Back to top](#bc)</div>

---

<a name="refs_gc-references-sharing"></a>
###_Sharing_
There are a lot of ways `References` can be shared but I'm only going to focus on three. 

If you initialize a variable `a` and set it equal to a varible `b` holding a `Reference`, `a` now holds that same `Reference` (or points to, whichever phrasing works better for you). Notice, we're back to the equality situation again.

Assignment has the same behavior, with the an added caveat. If we have a assign `a` to a variable `c` holding a `Reference`, and `c` is not equal to `b`, `a` now holds the same reference as `c`. To clarify with a diagram:

$$A_{Variable}\ => C_{Variable}\ =>\ C_{Reference}$$
$$B_{Variable}\ =>\ B_{Reference}$$

The takeaway is that `a` no longer equals `b`, as they both hold different `Reference`s, implying that assignments overwrite the previous `Reference` of a variable. For `a` and `b` to be equal again, they would have to be assigned to one another.

You can also pass references into functions. When doing so, the only difference from the above statement is that the `Reference` is stored in a variable local to the function. Otherwise, all same rules still apply. 

<div style="text-align:right;">[Back to top](#bc)</div>

---

<a name="refs_gc-references-deref"></a>
###_Dereferencing_
Usually an ID is for _accessing_ something, and that's no different in this case. A `Reference` usually points to a thing. Things, in the world of programming, are typically called `Object`s. 

$$Reference\ =>\ Object$$

This is also important because we're usually more interested in the `Object`. I'm referring to the act of obtaing the `Object` via its `Reference` as _derferencing_.

Thus, any variable holding a `Reference` may be _dereferenced_ to obtain the `Object`. `Object`s can be anyting like numbers, strings, arrays, classes, whatever.

I have yet to come across a language built on `References`, or a situation in such a language, that doesn't automatically _derefence_ when dealing with variables. The languages are pretty solid at differentiating between when the `Reference` or `Object` is needed. If anyone knows of examples otherwise, please let me know.

<div style="text-align:right;">[Back to top](#bc)</div>

---

<a name="refs_gc-references-null"></a>
###_Null_
A lot of languages support a special type of `Reference` that, if dereferenced, would point to nothing. There are variety of names for this `Reference` but it's usually referred to as the `Null`. 

Even though `Null` refers to nothing, it still behaves like any other `Reference`.

<div style="text-align:right;">[Back to top](#bc)</div>

---

<a name="refs_gc-references-count"></a>
###_Reference Counting_
If IDs are important for accessing things, then it's important to track who has the ID in question, even moreso if the ID is being passed around.

Enter, reference counting. Every `Reference`, except `Null`, has a `Reference Count`: the number of variables holding that `Reference`. Every time a `Reference` is passed to a variable, the corresponding `Reference Count` is incremented. Every time a variable switches to holding a different `Reference`, including `Null`, the `Reference Count` is decremented. 

And what makes `Null` so special that it gets no `Reference Count`? It doesn't refer to anything. So there really is no point in tracking the number of variables holding `Null`.

So to review:

- Increment `Reference Count` when a `Reference` is shared
- Decrement `Reference Count` when a `Reference` is overwritten

<div style="text-align:right;">[Back to top](#bc)</div>

---

<a name="refs_gc-garbage_collection"></a>
Garbage Collection
------------------
So what about `Reference`s with a `Reference Count` of 0? This implies that there's a `Reference` pointing to an `Object`, that isn't held by any variables, meaning, it's `Object` is not being used. Unusued `Object`s take up memory, so it's nice to get rid of them. This is the point of garbage collection. `Garbage Collection` looks at `Reference Counts`, and if a `Reference` has a count of 0, it's corresponding `Object` and the `Reference` are removed from memory.

<div style="text-align:right;">[Back to top](#bc)</div>

---

<a name="refs_gc-complicated"></a>
So Complicated! Why?
--------------------
Prior to `References` and `Garbage Collection`, you were responsible for rolling your own memory management. This meant **manually** making space in memory for variable; **manuaully** clearing that space when you were done; **manually** tracking which variables referenced sections of memory; **manually** a lot of other things and memory. 

And if you forget to clear memeory space when you're done, that's a memory leak. And if you clear memory prior to your variables using it, who knows what will happen. And if you write to the wrong space in memory, who know's what you just overwrote. And if you write to a reserved space in memory, have fun. It's an exhaustive list.

`References` and `Garbage Collection` were created as a solution to the above mess. As an abstraction for memory management, developers get the joys of creating `Objects` and variables fewer hassles. It comes with it's own set of quirks and inefficencies though, which I'm not going to cover.
